```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/01-04-2021.csv'
df = pd.read_csv(path)
```

```python
# 1. Write a Python program to display the first 5 rows from COVID-19 dataset. Also print the dataset information and check the missing values.
df.head()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1) | df.isna().any(axis=1)]
```

```python
# 2. Write a python program to replace all the missing values with “Tobefound” for province column.
df['Province_State'].replace(np.nan, "Tobefound", inplace = True)
df[['Province_State']].head()
```

```python
# 3. Find the Z-Score for column “Recovered”
from scipy import stats
df1 = df
df1.replace(np.nan, 0, inplace=True)
df.dropna(inplace=True)
df1['zscore_recovered']=stats.zscore(df["Recovered"])
```

```python
df1.loc[df1['zscore_recovered'].abs()>1]
```

```python
# Create bin values for “Recovered” column in the data set.
bins = np.linspace(min(df["Recovered"]), max(df["Recovered"]), 4)
group_names = ['Low', 'Medium', 'High']
df["Recovered-Binned"] = pd.cut(df["Recovered"], bins, labels=group_names, include_lowest = True)
df[['Recovered','Recovered-Binned']].head(20)
```

```python
# 5. Write a Python program to get the maximum number of confirmed, deaths, recovered and active cases of Novel Coronavirus (COVID-19) Country wise.
df.groupby(['Country_Region'])[['Confirmed']].max()
```

```python
# 6. Write a Python program to get the total number of confirmed deaths and recovered people of Novel Coronavirus (COVID-19) cases Country/Region - Province/State wise
df.groupby(['Country_Region', 'Province_State'])[['Confirmed']].max()
```

```python
df.groupby(['Country_Region', 'Province_State'])[['Deaths']].max()
```

```python
# 7. Write a Python program to create a plot (lines) of total deaths, confirmed, recovered and active cases Country wise
plt.figure(figsize=(20, 20))
plt.plot(df['Province_State'], df['Deaths'])
plt.plot(df['Province_State'], df['Confirmed'])
plt.plot(df['Province_State'], df['Recovered'])
plt.xticks(rotation='vertical')
```
